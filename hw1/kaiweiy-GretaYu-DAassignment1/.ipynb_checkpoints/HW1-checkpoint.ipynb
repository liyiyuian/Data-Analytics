### 1
Download historical daily weather data for France. For example, the analysis could be based on the weather in Paris by using: http://www.wunderground.com/history/airport/LFPO/  
Organize the data so that it can be saved as a csv file and load it into your computer. Fill any gaps in the data using linear interpolation.

```python
from pandas import read_csv
from pandas import datetime

weather = read_csv('Paris_weather_data_2017.csv', header=0, encoding="latin1",
                   na_values=['-'])

colDict = {}
for l in list(weather):
    colDict[l] = l.replace('\xa0', ' ')

weather.rename(index=str, columns=colDict, inplace=True)
weather.drop(['Events', 'high Gust Wind (km/h)'], axis=1, inplace=True)

print("Q1")
print("Before filling missing value")
print(weather.info())
weather
```

```python
weather.interpolate(inplace=True)

print("After filling missing value")
print(weather.info())
weather
```

### 3
Download historical daily electricity consumption data for France from:
http://clients.rte-france.com/lang/an/visiteurs/vie/vie_stats_conso_jour.jsp  
Save it as a csv file and load it into your computer.

```python
import pandas as pd
data_xls = pd.read_excel('Historique_consommation_JOUR_2017.xls', 'Sheet1', index_col=None)
data_xls.to_csv('electricity_consumption.csv', index=False)

energy = read_csv('electricity_consumption.csv', header=0)
energy.drop('Type de données', axis=1, inplace=True)
energy.dropna(inplace=True)
energy.reset_index(inplace=True, drop=True)
energy.rename(index=str, columns={'Energie journalière (MWh)':'Energy (MWh)'}, inplace=True)
```

### 4
Synchronize the dates corresponding to both time series and make a scatter plot of energy consumption against mean temperature.


```python
consumption = energy.loc[:,'Energy (MWh)']
meanTemp = weather.loc[:,'avg Temp. (°C)']

fig, ax = plt.subplots()
ax.scatter(meanTemp, consumption, alpha=0.8)

ax.set_xlabel('Mean temperature (°C)')
ax.set_ylabel('Energy consumption (MWh)')
ax.set_title('Temperature vs Energy')

ax.grid(True)
fig.tight_layout()

plt.show()
```

### 5
Fit a quadratic model to the energy versus temperature. Plot the quadratic fit as a line on top of the scatter plot.

```python
import numpy as np

quadCoeff = np.polyfit(meanTemp, consumption, 2)

xPlt = np.linspace(-5,30,100)
quad = quadCoeff[2] + xPlt*quadCoeff[1] + (xPlt**2)*quadCoeff[0]
print("Quadratic equation:\n y = {} * x^2 - {} * x + {}".format(quadCoeff[2],quadCoeff[1],quadCoeff[0]))
```

```python
fig, ax = plt.subplots()

ax.scatter(meanTemp, consumption, alpha=0.8)
ax.plot(xPlt, quad)

ax.set_xlabel('Mean temperature (°C)')
ax.set_ylabel('Energy consumption (MWh)')
ax.set_title('Temperature vs Energy')

ax.grid(True)
fig.tight_layout()

plt.show()
```

### 6
Based on the empirical analysis, what is the optimal temperature coinciding with minimal consumption? Use the quadratic fit and verify visually.

```python
from sympy import *
from sympy.solvers import solve

minCons = consumption.min()

x = symbols('x')
output = solve(quadCoeff[0]*(x**2) + quadCoeff[1]*x + quadCoeff[2], x)
output = sum(output)/2
outputY = quadCoeff[0]*(output**2) + quadCoeff[1]*output + quadCoeff[2]

print("Optimal temperature calculated by quadratic fit: {}°C".format(output))
```

```python
fig, ax = plt.subplots()

ax.scatter(meanTemp, consumption, alpha=0.8, label="")
ax.plot(xPlt, quad)
ax.scatter(meanTemp.loc[consumption.idxmin], consumption.loc[consumption.idxmin], c='r',label="min consumption")
ax.scatter(output, outputY, c='r', marker="x", label="optimal Temp. for min consumption")

ax.set_xlabel('Mean temperature (°C)')
ax.set_ylabel('Energy consumption (MWh)')
ax.set_title('Temperature vs Energy')
ax.legend()

ax.grid(True)
fig.tight_layout()

plt.show()

print("Optimal temperature from finding minimum consumption in data: {}°C".format(meanTemp.loc[consumption.idxmin]))
print("For the minimum consumption, the optimal temperature calculated by quadratic fit is offset from the real value by {}°C".format(round(output-meanTemp.loc[consumption.idxmin], 4)))
```

### 8
Increase the number of explanatory variables by also considering squared terms for each weather variable. Use a stepwise approach to obtain a new model. Which variables are selected? What is the new R2 value and is this an improvement?


```python
weatherSq = weather.drop('Date', axis=1)**2

colDict = {}
for l in list(weatherSq):
    colDict[l] = l + '_sq'

weatherSq.rename(index=str, columns=colDict, inplace=True)
weather_sq = weather.join(weatherSq)

weather_sq.info()
weather_sq
```

```python
def compareModels(mdl1,mdl2):
    if mdl1.rsqr > mdl2.rsqr:
        print("{} is not as good as {}. R-Squared drop by {}".format(mdl2.name, mdl1.name, mdl1.rsqr-mdl2.rsqr))
    elif mdl1.rsqr < mdl2.rsqr:
        print("{} performs better than {}. The R-Squared improved by {}".format(mdl2.name, mdl1.name, mdl2.rsqr-mdl1.rsqr))     
```

```python
mdl2 = Model("Model_2")
stepwise(weather_sq.drop('Date', axis=1), energy.drop('Date', axis=1), mdl2)
compareModels(mdl1,mdl2)
```

### 9 
Consider the day of the week effect by including dummy variables for the day of the week in the multivariate regression. Which days of the week are selected for the new model? What is the new R2 value and does this improve the model?

```python
# import time
from datetime import datetime

format = "%d/%m/%Y"

dayOfWeek = []
date = energy.loc[:,'Date']
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]


for i in range(len(date)):
    dayOfWeek.append(days[datetime.strptime(date[i], format).date().weekday()])

dum = pd.get_dummies(pd.Series(dayOfWeek))
weather_sq.reset_index(drop=True, inplace=True)
dum.reset_index(drop=True, inplace=True)

weather_wd = weather_sq.join(dum)
energy.reset_index(drop=True, inplace=True)

weather_wd.info()
weather_wd
```

```python
mdl3 = Model("Model_3")
stepwise(weather_wd.drop('Date', axis=1), energy.drop('Date', axis=1), mdl3)

compareModels(mdl2,mdl3)
```
